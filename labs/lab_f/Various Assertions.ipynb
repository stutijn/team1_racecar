```python
def clamp(value: float, min: float, max: float) -> float:
    if value < min:
        return min
    elif value > max:
        return max
    else:
        return value

assert clamp(5, -1, 1)

def remap_range(val: float, old_min: float, old_max: float, new_min: float, new_max: float) -> float:
    old_range = old_max - old_min
    new_range = new_max - new_min
    return new_range * (float(val - old_min) / float(old_range)) + new_min

assert remap_range(50, -100, 100, -1, 1)
remap_range(.54, -1, 1, 0, 1023)
remap_range(64, 0, 100, 0, 200)
remap_range()
```
