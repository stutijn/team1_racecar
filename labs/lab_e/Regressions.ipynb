```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# (x, y) = (time (s), distance (cm))
data = [(2, 100), (4, 200), (6, 300), (8, 400), (10, 500)]

# Transpose data into two elements each containing x and y values
data_t = np.transpose(data)

# Determine linear regression beta values
b_val = np.polyfit(data_t[0], data_t[1], 1)
x = np.linspace(0, 10, 100) # Generate discrete plot points
y = b_val[0] * x + b_val[1] # Generate function
plt.plot(x, y, label='regression') # Plot regression
plt.scatter(data_t[0], data_t[1], label='raw data', color='red') # Plot original data points

# Graph settings and features
plt.legend()
plt.grid(True, linestyle=':')
plt.title("Relationship between Distance and Time")
plt.xlabel("Time (s)")
plt.ylabel("Distance (cm)")
plt.show()
```

```python
def inv_function(x, a, b):
    return a / x + b

#Dataset for distance vs area dataset
x_data = np.array([25, 40, 80, 120, 160, 200])
y_data = np.array([28000, 16249, 4716, 2187, 1238, 802])

# Linspace x for running the functions
x = np.linspace(2, 200, 100) # Generate discrete plot points

# Use curve-fit to fit the model to the data
params, covariance = curve_fit(inv_function, x_data, y_data)
print(f"Parameters: {params} || Covariance: {covariance}")

a_fit, b_fit = params

y = a_fit / x + b_fit

plt.scatter(x_data, y_data, label='raw data', color='red')
plt.plot(x, y, label='regression')


# Plot settings
plt.legend()
plt.grid(True, linestyle=':')
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.title("Relationship between Contour Area and Distance")
plt.xlabel("Distance (cm)")
plt.ylabel("Contour Area (px)")
plt.show()

inv_function(50, a_fit, b_fit)
```

```python


def inv_function(x, a, b, c):
    return a / (x + b) + c

#Dataset for distance vs area dataset
x_data = np.array([1, 2, 3, 4, 5, 10, 15, 20]) # time in hours
y_data = np.array([535, 740, 850, 920, 975, 1090, 1135, 1160]) # distance traveled in meters

# Linspace x for running the functions
x = np.linspace(2, 200, 100) # Generate discrete plot points

# Use curve-fit to fit the model to the data
params, covariance = curve_fit(inv_function, x_data, y_data)
print(f"Parameters: {params} || Covariance: {covariance}")

a_fit, b_fit, c_fit = params

y = a_fit / (x + b_fit) + c_fit

plt.scatter(x_data, y_data, label='raw data', color='red')
plt.plot(x, y, label='regression')


# Plot settings
plt.legend()
plt.grid(True, linestyle=':')
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.title("Relationship between Distance traveled and Time")
plt.xlabel("Time (hours)")
plt.ylabel("Distance traveled (meters)")
plt.show()

inv_function(3.4, a_fit, b_fit, c_fit)
```
